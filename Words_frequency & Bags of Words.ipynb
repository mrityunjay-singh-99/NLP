```python
import numpy as np
import pandas as pd
import string,re
from nltk.tokenize import word_tokenize
from collections import Counter

#preprocessing text
def preprocess_text(text1):
    #convert text in lowercase
    text1 = text1.lower()
    #split the word
    words = word_tokenize(text1)
    
    #remove punctuation
    re_punch = re.compile("[%s]"%re.escape(string.punctuation))
    words = [re_punch.sub(" ", w) for w in words]
    words = [w for w in words if w.isalnum()]

    return words
    
words = preprocess_text(text)

#count frequency of unique words
words_count = Counter(words)
print("frequency_words")
for words,counts in words_count.items():
    print(f"{words}:{counts}")

```

```python
#create bags of matrixes
unique_words = sorted(words_count.keys())

#list
blow_matrix = []

# Split the text into sentences
sentences = [preprocess_text(sentence) for sentence in text.split(".") if text.strip()]

# Create the BoW matrix for each sentence
for sentence in sentences:
    row = [sentence.count(w) for w in unique_words]
    blob_matrix.append(row)

#convert a pandas dataframe 
blow_df = pd.DataFrame(blob_matrix, columns = unique_words)

blow_df
```
